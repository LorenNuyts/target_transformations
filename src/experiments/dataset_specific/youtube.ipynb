# Find time that video has been online

## What is the maximum publish date? That will be our reference date

```python
df_publish = data[['publish_year', 'publish_month', 'publish_day']]
df_publish.rename(columns={'publish_year': 'year', 'publish_month': 'month', 'publish_day': 'day'}, inplace=True)
df_publish = df_publish.astype(int)
publish_dates = pd.to_datetime(df_publish,format='%Y%m%d')
max_date = publish_dates.max()
```

## Convert publish dates to time online

```python
time_online = max_date - publish_dates
data['time_online'] = time_online.dt.days
```
